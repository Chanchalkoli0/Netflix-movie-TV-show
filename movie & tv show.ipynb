```python
df[['director','date_added']] = df[['director','date_added']].fillna("Unknown")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['cast'], inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['country'] = df['country'].fillna("unknown")
df.head()
```

```python
df.isnull().sum()
```

```python
df[['rating','duration']] = df[['rating','duration']].fillna("NA")
df.head()
```

```python
df.isnull().sum()
```

#convert all country names in lowercase

```python
df['country'] = df['country'].str.lower().str.strip()
df.head()
```

```python
#replacing Male and Female with M and F respectively.
```

```python
df['cast'] = df['cast'].str.replace("M", "Male").str.replace("F", "Female")
df.head()
```

#conversion of 'date_added' column to datetime as (yyyy-mm-dd)

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
df.head()
```

#converting all columns name with lowercase and also removing spaces with underscores

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
#converting data type of release_year
```

```python
df['release_year'] = df['release_year'].astype(str)
df.dtypes
```
